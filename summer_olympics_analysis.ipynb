```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Summer.csv')

# 1. In how many cities Summer Olympics is held so far?

total_cities = []
for city in data['City'].unique():
    total_cities.append([city, len(data[data['City']==city])])

pd.DataFrame(total_cities,columns = ['city','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'city', y = 'freq', kind = 'bar', figsize = (5,5))
```

```python
# 2. Which sport is having most number of Gold Medals so far? (Top 5)
medals = []
for sport in data['Sport'].unique():
    medals.append([sport, len(data[data['Medal'] == 'Gold'])])

top_5_sport = medals[:5]
pd.DataFrame(top_5_sport,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (5,5))
```

```python
# 3. Which sport is having most number of medals so far? (Top 5)

all_medals = []
for sport in data['Sport'].unique():
    all_medals.append([sport, len(data['Medal'])])

top_5_medals_sport = all_medals[:5]

pd.DataFrame(top_5_medals_sport,columns = ['sport','medal']).sort_values(by='medal', ascending=False).head().plot(x = 'sport', y = 'medal', kind = 'bar', figsize = (5,5))
```

```python
# 4. Which player has won most number of medals? (Top 5)

player_medals = []
for player in data['Athlete'].unique():
    player_medals.append([player, len(data['Medal'])])

top_5_player  = player_medals[:5]
print('Ans: ')
for name in top_5_player:
    print(name[0])
pd.DataFrame(top_5_player,columns = ['name','medals']).sort_values(by='medals', ascending=False).head().plot(x = 'name', y = 'medals', kind = 'bar', figsize = (5,5))
```

```python
player_medals_gold = []
for player in data['Athlete'].unique():
    player_medals.append([player, len(data[data['Medal'] == 'Gold'])])

top_5_gold  = player_medals_gold[:5]
print('Ans: ')
for name in top_5_gold:
    print(name[0])
pd.DataFrame(top_5_gold,columns = ['name','medals']).sort_values(by='medals', ascending=False).head().plot(x = 'name', y = 'medals', kind = 'bar', figsize = (5,5))
```

```python
# Which sport is having most female Gold Medalists?
f_player = []
for sport in data["Sport"].unique():
    f_player.append([sport, len(data[data['Gender'] == 'Women'])])

top_5_f = f_player[:5]
pd.DataFrame(top_5_f,columns = ['sport','count']).sort_values(by='count', ascending=False).head().plot(x = 'sport', y = 'count', kind = 'bar', figsize = (5,5))
```
